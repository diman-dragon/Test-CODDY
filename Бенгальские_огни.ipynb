У Игоря есть с1 бенгальских огней. Когда Игорь тратит один огонек, сначала он сверкает два часа, а затем тухнет. Игорь умный парень и из b1 потухших огоньков можно сделать 2 новых бенгальских огня, которые можно зажечь.

Теперь Игорю интересно, сколько часов будет гореть огонек, если он будет действовать оптимальным образом. 

Для решения этой задачи нужно учитывать два этапа: использование изначальных бенгальских огней и создание новых из потухших. Важно оптимально использовать каждый потухший огонек, чтобы максимально продлить общую продолжительность горения.

Вот пошаговое решение на Python:

Используем изначальные бенгальские огни.  
Каждый раз, когда у нас есть достаточно потухших огоньков, создаем новые огни и продолжаем их использовать.  
Повторяем процесс, пока это возможно.  

```python
import matplotlib.pyplot as plt

def total_burn_hours(c1, b1):
    total_hours = 0
    spent_ogonki = 0
    ogonki_history = [c1]  # Записываем историю количества огней
    ignitions = []  # Список для хранения моментов зажигания

    while c1 > 0:
        # Тратим все доступные огоньки
        total_hours += c1 * 2
        spent_ogonki += c1
        c1 = 0
        
        # Создаем новые огоньки из потухших
        c1 = spent_ogonki // b1
        spent_ogonki = spent_ogonki % b1
        
        ogonki_history.append(c1)  # Записываем количество огней после создания новых
        ignitions.append(len(ogonki_history) - 1)  # Записываем момент зажигания

    return total_hours, ogonki_history, ignitions

# Пример использования:
c1 = 300  # начальное количество бенгальских огней
b1 = 2   # количество потухших огоньков, необходимых для создания 2 новых

total_hours, ogonki_history, ignitions = total_burn_hours(c1, b1)
print("Общее время горения:", total_hours)
print("История количества огней:", ogonki_history)
print("Моменты зажигания огней:", ignitions)

# Визуализируем процесс
plt.plot(range(len(ogonki_history)), ogonki_history, marker='o')
plt.title('Изменение количества бенгальских огней')
plt.xlabel('Шаги')
plt.ylabel('Количество огней')
plt.grid(True)

# Добавляем вертикальные линии для моментов зажигания
for ignition in ignitions:
    plt.axvline(x=ignition, color='r', linestyle='--', linewidth=0.5)

plt.show()
```
